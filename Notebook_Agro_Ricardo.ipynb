## ¿Qué es Python?

Python es un lenguaje de programación de **alto nivel**, lo que en términos simples significa que es muy parecido a leer inglés. Es famoso por ser:

* **Legible:** Su sintaxis es limpia, ideal para ingenieros que no son especialistas en sistemas.
* **Versátil:** Sirve para analizar datos, controlar sensores, automatizar máquinas o crear inteligencia artificial.
* **Comunitario:** Tiene miles de librerías gratuitas que ya hacen el trabajo pesado por ti (procesamiento de imágenes, estadística, etc.).

## ¿Por qué es importante en la Ingeniería Agroindustrial?

La agroindustria moderna ya no solo trata de transformar materia prima, sino de gestionar **información**. Python es clave para:

1. **Agricultura de Precisión:** Analizar datos de suelos y clima para optimizar el riego y fertilización.
2. **Control de Procesos:** Monitorear variables como temperatura y pH en tiempo real durante la transformación de alimentos.
3. **Optimización Logística:** Calcular las rutas más rápidas para la distribución o gestionar inventarios para reducir mermas.
4. **Control de Calidad:** Usar visión artificial para clasificar frutas o detectar defectos automáticamente.

## 3 Ejemplos Sencillos y Prácticos

Aquí tienes tres aplicaciones básicas que podrías empezar a usar hoy mismo:

### 1. Cálculo del Índice de Madurez (Relación Grados Brix/Acidez)

En la industria de jugos o frutas, este índice determina el momento óptimo de cosecha o procesamiento.

```
# Datos de la muestra
grados_brix = 12.5
acidez_titulable = 0.8

# Cálculo del índice
indice_madurez = grados_brix / acidez_titulable

print(f"El índice de madurez es: {indice_madurez}")

if indice_madurez > 15:
    print("Estado: Fruta muy dulce / Madura")
else:
    print("Estado: Fruta apta para procesamiento o requiere más tiempo")
```

### 2. Control de Humedad en Secado

Un pequeño script para verificar si un lote de grano ya alcanzó la humedad comercial (por ejemplo, 13%).

```
humedad_actual = 18.5  # Dato que vendría de un sensor
humedad_objetivo = 13.0

if humedad_actual > humedad_objetivo:
    diferencia = humedad_actual - humedad_objetivo
    print(f"Alerta: El grano sigue húmedo. Falta reducir un {diferencia}%")
else:
    print("Proceso terminado: Humedad óptima alcanzada.")
```

### 3. Estimación de Rendimiento de Extracción

Útil en la extracción de aceites esenciales o almidones para medir la eficiencia de la planta.

```
peso_materia_prima = 500  # kg
peso_producto_final = 45   # kg

rendimiento = (peso_producto_final / peso_materia_prima) * 100

print(f"El rendimiento del proceso es del {rendimiento}%")

if rendimiento < 8:
    print("Revisar eficiencia de la maquinaria: Rendimiento por debajo del estándar.")
```
