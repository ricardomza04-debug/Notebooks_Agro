# **Índice de temperatura y humedad (producción intensiva)**

$$ IPH = 0.72 * (T_{bs} + T_{bh}) + 40.6 $$

Donde:

1. $T_{bs}$: Temperatura bulbo seco ($ºC$)

2. $T_{bh}$: Temperatura de bulbo humedo ($ºC$)

Para la correlacion de los datos de la especie bovina en la produccion de leche se recomienda utilizar los limites propuestos por National weather service:

+ Si $ITH <= 70$ ambiente no estresante
+ Si $ITH > 70 and ITH <= 78$ situacion de alerta por estres termico
+ Si $ITH > 78 and ITH <= 83$ peligro por estres termico
+ Si $ITH > 83$ situacion critica

```python
# ENTRADA: Temperatura De Bulto Seco (tbs,grados centigrados)
# Temperatura De Bulto Humedo (tbh,grados centigrados)
tbs = float(input("ingrese la temperatura de bulto seco, grados centigrados: "))
tbh = float(input("ingrese la temperatura de bulto humedo, grados centigrados: "))

# PROCESO: calculo de ith
ith = 0.72 * (tbs + tbh) + 40.6

# SALIDA: Ith
if ith <= 70:
  print("ambiente no estresante")
elif ith > 70 and ith <= 78:
  print("situacion de alerta por estres termico")
elif ith > 78 and ith <= 83:
  print("peligro por estres termico")
elif ith > 83:
  print("situacion critica")
```
